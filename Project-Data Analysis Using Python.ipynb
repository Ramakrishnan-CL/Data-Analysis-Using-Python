```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("C:/Users/ramak/OneDrive/Desktop/Python projects/sales_dataset.csv") #load dataset
```

```python
data.head() #print first few rows to check
```

```python
data['Total'].sum() #finding total sum
```

```python
data.groupby('Branch')['Total'].sum().sort_values(ascending=False) # finding total sum by each branch
```

```python
(data.groupby('Category')['Total'].sum() / data['Total'].sum()) * 100 #finding % of total sum contributed by each category
```

```python
data.groupby('Category')['Total'].sum().sort_values(ascending=False) #finding the category which contributed to highest revenue
```

```python
data.groupby('Gender')['Total'].sum().sort_values(ascending=False) #total sum by gender
```

```python
data['Date'] = pd.to_datetime(data['Date'])
monthly_sales = data.resample('ME', on='Date')['Total'].sum()
monthly_sales.plot(kind='line', marker='o', color='green')
plt.xlabel('Date')
plt.ylabel('Total Sales')
plt.title('Total Sales Over Time')
plt.show() #Representation of Total sales over time using line graph
```

```python
plt.scatter(data['Unit Price'], data['Total'], color='red')
plt.xlabel('Unit Price')
plt.ylabel('Total Sales')
plt.title('Scatter Plot of Unit Price vs. Total Sales')
plt.show() #Representating unit price vs Total Sales using Scatter Plot
```

```python
data.groupby('Payment Type')['Total'].sum().plot(kind='bar')
plt.title('Total Sales by Payment Type')
plt.ylabel('Total Sales')
plt.show() #Representing total sales per branch using Bar Graph
```

```python
plt.hist(data['Total'], bins=10, color='pink', edgecolor='black')
plt.xlabel('Total Sales')
plt.ylabel('Frequency')
plt.title('Histogram of Total Sales')
plt.show() #Representing Total sales vs Frequency using Histogram
```
